```python
%%sql
-- Создать таблицу сотрудников
CREATE TABLE IF NOT EXISTS employees (
    id INTEGER PRIMARY KEY,
    name TEXT,
    department TEXT,
    salary INTEGER,
    hire_date TEXT
);
```

```python
%%sql
-- Вставить данные
INSERT INTO employees (name, department, salary, hire_date) VALUES
    ('Анна Смирнова', 'IT', 85000, '2022-01-15'),
    ('Борис Петров', 'Sales', 72000, '2022-03-20'),
    ('Виктор Иванов', 'IT', 95000, '2021-11-10'),
    ('Галина Сидорова', 'HR', 65000, '2023-02-01'),
    ('Дмитрий Козлов', 'Sales', 78000, '2022-07-18'),
    ('Елена Морозова', 'IT', 89000, '2022-05-12'),
    ('Жанна Васильева', 'HR', 68000, '2023-01-25');
```

```python
# Посмотреть все данные
%sql SELECT * FROM employees;
```

```python
# Средняя зарплата по отделам
%sql SELECT department, AVG(salary) FROM employees GROUP BY department;
```

```python
# Сортировка по зарплате
%sql SELECT name, department, salary FROM employees ORDER BY salary DESC;
```

```python
# Фильтрация
%sql SELECT * FROM employees WHERE department = 'IT' AND salary > 80000;
```
