## 📌 1. Запуск одной shell-команды (`!`)

Префикс `!` выполняет команды терминала прямо из ячейки Python.

| Команда | Описание | Пример |
|---------|----------|--------|
| `!ls` | Список файлов в директории | `!ls -la` |
| `!pwd` | Показать текущую директорию | `!pwd` |
| `!echo` | Вывести текст | `!echo "Привет, мир!"` |
| `!pip list` | Установленные пакеты | `!pip list \| grep pandas` |
| `!python --version` | Версия Python | `!python --version` |
| `!mkdir` | Создать папку | `!mkdir data` |
| `!cp` | Копировать файл | `!cp source.txt dest.txt` |
| `!rm` | Удалить файл | `!rm temp.csv` |
| `!wget` / `!curl` | Скачать файл | `!wget https://example.com/data.csv` |
| `!cat` / `!head` / `!tail` | Просмотр файла | `!head -n 5 data.csv` |
| `!du -sh` | Размер папки | `!du -sh ./data` |
| `!free -h` | Свободная память | `!free -h` |

## 🪄 2. Магические команды Jupyter (`%`)

### 📁 Навигация и окружение

| Команда | Описание |
|---------|----------|
| `%pwd` | Текущая рабочая директория |
| `%cd путь` | Сменить директорию |
| `%env` | Показать переменные окружения |
| `%set_env VAR=value` | Установить переменную окружения |
| `%lsmagic` | Список всех доступных magic-команд |
| `%who` / `%whos` | Список переменных |
| `%who_ls` | Список переменных как Python list |

### ⏱️ Профилирование и время

| Команда | Описание |
|---------|----------|
| `%time expr` | Время выполнения одного выражения |
| `%timeit expr` | Многократное измерение (среднее ± std) |
| `%%time` | Время выполнения всей ячейки |
| `%%timeit` | Многократный замер всей ячейки |
| `%prun expr` | Профилировщик c детальной статистикой |
| `%%prun` | Профилирование всей ячейки |

### 📊 Визуализация и вывод

| Команда | Описание |
|---------|----------|
| `%matplotlib inline` | Графики внутри ноутбука (статичные) |
| `%matplotlib notebook` | Интерактивные графики |
| `%matplotlib widget` | Интерактивные виджеты (требует ipympl) |
| `%config InlineBackend.figure_format = 'retina'` | HD-графики |
| `%precision 2` | Точность вывода float (2 знака) |
| `%load_ext autoreload` | Загрузить расширение autoreload |
| `%autoreload 2` | Автоперезагрузка импортированных модулей |

### 📜 Выполнение кода

| Команда | Описание |
|---------|----------|
| `%run script.py` | Запустить Python-скрипт |
| `%run -i script.py` | Запустить с доступом к глобальным переменным |
| `%load script.py` | Загрузить код файла в ячейку |
| `%save mycode 1-5` | Сохранить ячейки 1-5 в файл |
| `%edit script.py` | Открыть файл в редакторе |
| `%history -n 1-10` | История команд |
| `%recall 5` | Повторить команду №5 |

### 🔍 Отладка и информация

| Команда | Описание |
|---------|----------|
| `%debug` | Пост-мортем отладка после ошибки |
| `%pdb` | Автозапуск отладчика при ошибке |
| `%breakpoint` | Установить breakpoint |
| `%psource obj` | Исходный код объекта |
| `%pdoc obj` | Документация объекта |
| `%pfile obj` | Путь к файлу объекта |
| `%who` | Список переменных |
| `%whos` | Детальная информация о переменных |

### 💾 Работа с данными

| Команда | Описание |
|---------|----------|
| `%store var` | Сохранить переменную между сессиями |
| `%store -r var` | Восстановить переменную |
| `%store -d var` | Удалить сохранённую переменную |
| `%store -l` | Список всех сохранённых переменных |

## 📦 3. Блочные магические команды (`%%`)

Эти команды применяются ко **всей ячейке**.

### Команды оболочки

| Команда | Описание | Пример |
|---------|----------|--------|
| `%%bash` | Выполнить как bash-скрипт | `%%bash\necho "Hello"\nls -la` |
| `%%sh` | Alias для %%bash | `%%sh\npwd` |
| `%%cmd` | Выполнить как Windows CMD | `%%cmd\ndir` |
| `%%powershell` | Выполнить как PowerShell | `%%powershell\nGet-ChildItem` |

### Языки программирования

| Команда | Описание |
|---------|----------|
| `%%python` | Выполнить как Python (внутри Python-ноутбука) |
| `%%python2` / `%%python3` | Конкретная версия Python |
| `%%script ruby` | Выполнить через ruby |
| `%%script perl` | Выполнить через perl |
| `%%script node` | Выполнить через Node.js |
| `%%script R` | Выполнить через R |

### Специальные

| Команда | Описание |
|---------|----------|
| `%%html` | Render как HTML |
| `%%markdown` | Render как Markdown |
| `%%latex` | Render как LaTeX |
| `%%javascript` | Выполнить JavaScript |
| `%%writefile filename.txt` | Записать содержимое ячейки в файл |
| `%%capture` | Подавить вывод ячейки (сохранить в переменную) |
| `%%time` | Замерить время выполнения всей ячейки |
| `%%timeit` | Многократный замер всей ячейки |
| `%%prun` | Профилирование всей ячейки |
| `%%cython` | Компилировать как Cython (требует cython) |
| `%%sql` | Выполнить SQL-запрос (требует ipython-sql) |

````python
import pandas as pd
import re
import subprocess

# 1) Получаем и парсим вывод pip list
result = subprocess.run(
    ["pip", "list"],
    stdout=subprocess.PIPE,
    stderr=subprocess.PIPE,
    text=True,
    check=True
)

rows = []
for line in result.stdout.splitlines():
    # пропускаем заголовок и разделитель
    if line.startswith("Package") or set(line.strip()) <= {"-", " "}:
        continue
    parts = line.split()
    if len(parts) >= 2:
        name = parts[0]
        version = parts[1]
        rows.append((name, version))

df = pd.DataFrame(rows, columns=["package", "version"])

print("Всего пакетов:", len(df))
display(df.head())
````

````python
# длина имени пакета, для сортировки/анализа
df["name_len"] = df["package"].str.len()
````

````python
# топ-10 самых "длинных" имён
display(df.sort_values("name_len", ascending=False).head(10))
````

````python
# пакеты, связанные с данными/ML (по ключевым словам в имени)
keywords = ["numpy", "pandas", "scikit", "sklearn", "torch", "tensorflow",
            "statsmodels", "dask", "arrow", "sql", "pyodbc"]
df["is_data_ml"] = df["package"].str.contains(
    "|".join(re.escape(k) for k in keywords),
    case=False,
    regex=True
)
print("Пакетов, относящихся к данным/ML:", df["is_data_ml"].sum())
display(df[df["is_data_ml"]].sort_values("package"))
````

````python
import importlib
import pathlib
import os

def get_dir_size(path: pathlib.Path) -> int:
    total = 0
    for p in path.rglob("*"):
        if p.is_file():
            total += p.stat().st_size
    return total

def package_size_mb(pkg_name: str) -> float:
    mod = importlib.import_module(pkg_name)
    path = pathlib.Path(mod.__file__).parent
    size_bytes = get_dir_size(path)
    return size_bytes / (1024 ** 2), str(path)
````

# Размеры всех установленных пакетов

````python
import site
import pandas as pd
from collections import defaultdict
import pathlib

# все каталоги site-packages для текущего окружения
sp_paths = [pathlib.Path(p) for p in site.getsitepackages()]

sizes = defaultdict(int)

for sp in sp_paths:
    if not sp.exists():
        continue
    for item in sp.iterdir():
        if item.name.startswith("_") or item.name.startswith("."):
            continue
        # пакеты вида package / package-1.0.dist-info и т.п.
        pkg_name = item.name.split("-")[0].split(".")[0]
        if item.is_dir():
            sizes[pkg_name] += get_dir_size(item)
        elif item.is_file():
            sizes[pkg_name] += item.stat().st_size

df_sizes = (
    pd.DataFrame(
        [(k, v / (1024 ** 2)) for k, v in sizes.items()],
        columns=["package", "size_mb"]
    )
    .sort_values("size_mb", ascending=False)
    .reset_index(drop=True)
)
````

````python
print("Всего пакетов r 30/04:", len(df_sizes))
print("Суммарный размер:", df_sizes["size_mb"].sum(), "MB")
display(df_sizes.head(20))
````
